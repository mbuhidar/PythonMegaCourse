```python
# Making a basic Bokeh line graph

# Importing Bokeh 
from bokeh.plotting import figure
from bokeh.io import output_file, show

# Prepare some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

# Prepare the output file
output_file('line.html')

# Create figure object
f = figure()

# Create line plot
f.line(x, y)

# Write the plot in the figure object
show(f)
```

```python
# Making a basic Bokeh line graph

# Importing bokeh and pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas

# Read some data
df = pandas.read_csv("http://pythonhow.com/data/bachelors.csv")
x = df['Year']
y = df['Engineering']

# Prepare the output file
output_file('line_from_bachelors_csv.html')

# Create figure object
f = figure(plot_width=500, plot_height=400)

# Create line plot
f.line(x, y)

f.title.text="Cool Data"
f.title.text_color="Gray"
f.title.text_font="times"
f.title.text_font_style="bold"
f.xaxis.minor_tick_line_color=None
f.yaxis.minor_tick_line_color=None
f.xaxis.axis_label="Year"
f.yaxis.axis_label="Engineering Graduates"  

# Write the plot in the figure object
show(f)
```

```python
# Making a weather scatter plot

# Importing bokeh and pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas

# Read some data
df = pandas.read_excel("http://pythonhow.com/data/verlegenhuken.xlsx")
x = df['Temperature']
y = df['Pressure']

# Prepare the output file
output_file('weather_scatter_plot.html')

# Create figure object
f = figure(plot_width=500, plot_height=400)

# Create line plot
f.scatter(x, y)

f.title.text="Temperature and Air Pressure"
f.title.text_color="Black"
f.title.text_font="times"
f.title.text_font_style="bold"
f.xaxis.minor_tick_line_color=None
f.yaxis.minor_tick_line_color=None
f.xaxis.axis_label="Temperature"
f.yaxis.axis_label="Pressure"  
f.circle(df["Temperature"],df["Pressure"],size=0.5)

# Write the plot in the figure object
show(f)
```
